````python
# prompt: sort end_activities berdasarkan frekuensinya

from collections import Counter

# Dapatkan daftar end activities dari log
end_activities_list = [trace[-1]['concept:name'] for trace in log_not_only_abstract]

# Hitung frekuensi setiap end activity
end_activities_counts = Counter(end_activities_list)

# Urutkan end activities berdasarkan frekuensi secara menurun
sorted_end_activities = dict(sorted(end_activities_counts.items(), key=lambda item: item[1], reverse=True))

print("End activities sorted by frequency:")
for activity, count in sorted_end_activities.items():
    print(f"{activity}: {count}")
````

````python
# prompt: pada log_not_only_abstract buang case variant yang frekuensinya kurang dari 2 menurut sorted_end_activities

# Mendapatkan semua varian dari log
variants = pm4py.get_variants(log_not_only_abstract)

# Mengidentifikasi varian yang berakhir dengan end activities dengan frekuensi < 2
variants_to_remove = []
for variant, traces in variants.items():
    # Dapatkan end activity dari varian (end activity adalah aktivitas terakhir dalam urutan)
    if traces: # Pastikan varian tidak kosong
        end_activity_of_variant = traces[0][-1]['concept:name']
        # Periksa apakah end activity ini ada di sorted_end_activities dan frekuensinya < 2
        if end_activity_of_variant in sorted_end_activities and sorted_end_activities[end_activity_of_variant] < 2:
            variants_to_remove.append(variant)

# Membuat log baru yang hanya berisi varian dengan frekuensi end activity >= 2
filtered_log_by_end_activity_frequency = EventLog()

for trace in log_not_only_abstract:
    # Buat string representasi varian dari trace saat ini
    trace_variant = pm4py.utils.variant_util.get_variant_representation(trace)
    # Jika varian ini tidak termasuk dalam varian yang akan dihapus, tambahkan ke log baru
    if trace_variant not in variants_to_remove:
        filtered_log_by_end_activity_frequency.append(trace)

print(f"Original number of traces: {len(log_not_only_abstract)}")
print(f"Number of traces after filtering by end activity frequency: {len(filtered_log_by_end_activity_frequency)}")

# Periksa end activities dari log yang sudah difilter untuk memastikan filter berfungsi
end_activities_filtered = pm4py.get_end_activities(filtered_log_by_end_activity_frequency)
print("\nEnd activities in the filtered log and their frequencies:")
for activity, count in end_activities_filtered.items():
    print(f"{activity}: {count}")

log_not_only_abstract = filtered_log_by_end_activity_frequency

# Anda bisa menyimpan log yang sudah difilter jika diperlukan
# pm4py.write_xes(log_not_only_abstract, '/content/drive/MyDrive/Colab Notebooks/ISNPINSA/log_tanpa_Abstract_Progress_2_was_Changed_to_end_activity_freq_filtered.xes')
# print("\nFiltered event log saved to log_tanpa_Abstract_Progress_2_was_Changed_to_end_activity_freq_filtered.xes")
````
